# Explorative Datenanalyse des Datensatzes stores.csv

```python
# Import der notwendigen Bibliotheken
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Konfiguration
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Einlesen der CSV-Datei
df = pd.read_csv('stores.csv')
```

## Statistische Ãœbersicht numerischer Spalten

## Kategorische Variablen analysieren

```python
print("\nðŸ”¹ Ãœbersicht kategorischer Spalten:")
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
for col in categorical_cols:
    print(f"\nSpalte: {col}")
    print(df[col].value_counts(dropna=False))
    sns.countplot(y=col, data=df, order=df[col].value_counts().index)
    plt.title(f"HÃ¤ufigkeit: {col}")
    plt.show()
```

## Eindeutigkeit prÃ¼fen

PrÃ¼fen, ob eine Location mehreren States zugeordnet ist

```python
multi_state = df.groupby('location')['state'].nunique()
multi_state_violations = multi_state[multi_state > 1]
print("ðŸ”¹ Locations mit mehr als einem State:")
print(multi_state_violations)
```

PrÃ¼fen, ob eine Location mehreren Urbanization-Werten zugeordnet ist

```python
multi_urban = df.groupby('location')['urbanization'].nunique()
multi_urban_violations = multi_urban[multi_urban > 1]
print("\nðŸ”¹ Locations mit mehr als einem Urbanization-Wert:")
print(multi_urban_violations)
```

Alle Zeilen mit Karlsruhe anzeigen

## PlausibilitÃ¤tsprÃ¼fung

PrÃ¼fen, ob SCO-EinfÃ¼hrung immer vor Store-ErÃ¶ffnung liegt

```python
# Umwandlung in datetime-Format
df['opening_date'] = pd.to_datetime(df['opening_date'], errors='coerce')
df['sco_introduction'] = pd.to_datetime(df['sco_introduction'], errors='coerce')

# Bedingung prÃ¼fen
invalid_dates = df[df['opening_date'] > df['sco_introduction']]

# Ergebnis anzeigen
print(f"ðŸ”¹ Anzahl Zeilen mit ungÃ¼ltiger Zeitlogik: {len(invalid_dates)}")
display(invalid_dates)
```

Zeitstrahl

```python
# Gruppieren nach Datum: Anzahl der ErÃ¶ffnungen pro Tag
openings_per_day = df['opening_date'].value_counts().sort_index()

# Gruppieren nach Datum: Anzahl der SCO-EinfÃ¼hrungen pro Tag
sco_per_day = df['sco_introduction'].value_counts().sort_index()

# Plot erstellen
fig, axes = plt.subplots(2, 1, figsize=(14, 10), sharex=False)

# Store Openings
axes[0].plot(openings_per_day.index, openings_per_day.values, marker='o')
axes[0].set_title('Store Openings Over Time')
axes[0].set_ylabel('Number of Openings')
axes[0].grid(True)

# SCO Introductions
axes[1].plot(sco_per_day.index, sco_per_day.values, marker='o', color='orange')
axes[1].set_title('Self-Checkout Introductions Over Time')
axes[1].set_ylabel('Number of SCO Introductions')
axes[1].grid(True)

plt.xlabel('Date')
plt.tight_layout()
plt.show()
```
